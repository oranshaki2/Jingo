```python
feature_names = tfidf.get_feature_names_out()
results = []

for idx, song in df.iterrows():
    song_vector = tfidf_matrix[idx]
    song_words = {}
    scores_by_category = {}
    
    for category, words in categories.items():
        score = 0
        found_words = []
        for word in words:
            if word in feature_names:
                word_idx = tfidf.vocabulary_.get(word)
                if word_idx is not None:
                    word_score = song_vector[0, word_idx]
                    if word_score > 0:
                        score += word_score
                        found_words.append(word)
        scores_by_category[category] = (score, found_words)
    
    # בחרי את הקטגוריה עם הציון הכי גבוה
    best_category = max(scores_by_category.items(), key=lambda x: x[1][0])
    
    results.append({
        "title": song["Song"],
        "artist": song["Artist"],
        "genre": song["Genre"],
        "category": best_category[0],
        "matched_words": best_category[1][1]
    })

results_df = pd.DataFrame(results)
print(results_df)
results_df.to_csv("../data/song_test_adam.csv", index=False, encoding='utf-8-sig')
```

```python
threshold = 0.1  # סף שאת קובעת - אפשר לשחק איתו

results = []

for idx, song in df.iterrows():
    song_vector = tfidf_matrix[idx]
    matched_categories = []
    
    for category, words in categories.items():
        score = 0
        found_words = []
        for word in words:
            if word in categories:
                word_idx = tfidf.vocabulary_.get(word)
                if word_idx is not None:
                    word_score = song_vector[0, word_idx]
                    if word_score > 0:
                        score += word_score
                        found_words.append(word)
        if score >= threshold:
            matched_categories.append({
                "category": category,
                "score": score,
                "matched_words": found_words
            })
    
    results.append({
        "title": song["Song"],
        "artist": song["Artist"],
        "genre": song["Genre"],
        "categories": matched_categories
    })

results_df = pd.DataFrame(results)
print(results_df.head())
results_df.to_csv("../data/song_test_adam.csv", index=False, encoding='utf-8-sig')
```
